```python
f = open('psalmen001Raw.txt', 'r')

lines = []

for line in f:
    lines.append(line)

f.close()
```

```python
import re

data = []

for line in lines:
    data_in_line = re.split(';',line)
    data.append(data_in_line)
```

```python
conversion_to_utf8 = {'>': "\u05D0",
                      'B': "\u05D1",
                      'G': "\u05D2",
                      'D': "\u05D3",
                      'H': "\u05D4",
                      'W': "\u05D5",
                      'Z': "\u05D6",
                      'X': "\u05D7",
                      'V': "\u05D8",
                      'J': "\u05D9",
                      'K': "\u05DB",
                      'L': "\u05DC",
                      'M': "\u05DE",
                      'N': "\u05E0",
                      'S': "\u05E1",
                      '<': "\u05E2",
                      'P': "\u05E4",
                      'Y': "\u05E6",
                      'Q': "\u05E7",
                      'R': "\u05E8",
                      'C': "\u05E9\u05C1",
                      'F': "\u05E9\u05C2",
                      'T': "\u05EA",
                      'p': "\u05E3",
                      'm': "\u05DD",
                      'n': "\u05DF",
                      'k': "\u05DA",
                      'y': "\u05E5",
                      '_': "-",
                      '-': "-",
                      '[': "[",
                      ']': "]",
                      '/': "/",
                      ' ': " ",
                      '\0': ""}

def set_word_to_unicode(sp):
    result = ""
    for c in sp:
        result += (conversion_to_utf8[c])
    return result
        
def analyze_phrase(subphrases):
    phrasesNew = ""

    if (len(subphrases) > 1):
        i = 0
        while (i < (len(subphrases) - 1)):
            phrasesNew += (set_word_to_unicode(subphrases[i]) + " ")
            i += 1
        label = subphrases[len(subphrases) - 1]
        phrasesNew += label + " "
    else:
        sf = re.split('-', subphrases[0])
        phrasesNew += set_word_to_unicode(sf[0]) + "-" + sf[1]
    return phrasesNew

def analyze_phrases(phrases):
    final_result = ""
    i = 0
    for phrase in reversed(phrases):
        final_result += analyze_phrase(phrase.split())
        if i > 0:
            final_result = final_result.strip() + "] "
        i += 1
    return final_result

def set_to_unicode(transcription, max_line_length):
    result = ""
    start_of_line = ""
    i = 0
    for c in transcription:
        if c == '[':
            break
        start_of_line += c
        i += 1
    rest_of_line = transcription[i:]
    whitespace = (max_line_length - len(transcription)) * " "
    result += whitespace + analyze_phrases(rest_of_line.split('] ')) + (''.join(reversed(start_of_line)))
    return result

def calculate_max_length(data):
    result = 0
    for d in data:
        transcription = d[19]
        if len(transcription) > result:
            result = len(transcription)
    return result
        
def analyze_transcription(data):
    unicode_lines = []
    max_line_length = calculate_max_length(data)
    for d in data:
        transcription = d[19]
        print (transcription)
        unicode_heb = set_to_unicode(transcription, max_line_length)
        unicode_lines.append(unicode_heb)
    return unicode_lines

unicode_lines = analyze_transcription(data)
for line in unicode_lines:
    print (line)
```

```python
from IPython.core.display import HTML

def get_hebrew_in_unicode(line):
    unicode_line_temp = line.replace("<", "&lt;")
    unicode_line_temp2 = unicode_line_temp.replace(">", "&gt;")
    return (unicode_line_temp2.replace(" ", "&nbsp;"))

def find_factors(d):
    if d[8] != "":
        if d[7] == "non-vol.":
            return "Morphological marking as " + d[8] + " overrules non-volitive default function"
        else:
            return "Morphological marking as " + d[8] + " confirms " + d[7] + " default function"
    elif d[10] != "":
        if d[10] == d[9]:
            if d[10] == "vol.pair":
                return "The clause forms a volitive pair together with its volitive daughter clause. The non-vol. default function is overridden by a volitive one."
            else:
                return "Inheritance of " + d[9] + " is possible and does indeed take place. The " + d[7] + " default function is overridden."
        else:
            if d[10] == "vol.pair":
                return "The clause could have formed a volitive pair together with its volitive daughter clause, but this is not the case. The non-vol. default function is not overridden."
            else:
                return "Inheritance of " + d[9] + " is possible, but does not take place. The " + d[7] + " default function is not overridden."
    elif d[12] != "":
        if d[12] == "attr." or d[12] == "constit.":
            "This 0-yiqtol clause is a " + d[12] + " clause and therefore does not fulfill its default " + d[7] + " function."
        elif d[12] == "precDrSp":
            "This 0-yiqtol clause introduces the preceding direct speech section, which should be seen as the 0-yiqtol verb's object. Therefore the clause does not fulfill its " + d[7] + " default volitive function."
        elif d[12] == "narr":
            "This 0-yiqtol clause is embedded in a narrative domain of communication and therefore does not fulfill its " + d[7] + "  default volitive function."
        elif d[12] == "prosp":
            "This 0-yiqtol clause is embedded in a prospective domain of communication and therefore does not fulfill its " + d[7] + "  default volitive function."
        elif d[12] == "inhExpSu":
            "This 0-yiqtol clause inherits its mother's explicit subject, should therefore be reanalyzed as an <X->yiqtol clause and so does not fulfill its " + d[7] + "  default volitive function."
        else:
            "This yiqtol clause inherits the multiple-duty modifier " + set_word_to_unicode(d[11]) + " and so does not fulfill its " + d[7] + "  default volitive function."
    else:
        return ""
            
def print_html():
    style_th = '<th style="padding:0px 15px 0px 0px;border-style:hidden;text-align:center">'
    style_td = '<td style="padding:0px 15px 0px 0px;border-style:hidden;">'
    style_unicode = '<td style="padding:0px 15px 0px 0px;border-style:hidden;text-align:right;font-family:SBL Hebrew;">'
    

    total = '<table style="border-style:hidden;">'
    head = '<tr style="border-style:hidden;">' + style_th + "Vs" + '</th>' + style_th + "\u00A7" + '</th>' + style_th + "Hebrew text" + '</th>' + style_th + "ClTp" + '</th>' + style_th + "CAR" + '</th>' + style_th + "DefFu" + '</th>' + style_th + "!" + '</th>' + style_th + "FinalFu" + '</th>' + style_th + "MDMod" + '</th>' + style_th + "Type" + '</th>' + style_th + "Lev" + '</th>' + style_th + "Persp" + '</th>' + style_th + "DiscFunction" + '</th>' + style_th + "Domain" + '</th>' + style_th + "#Pat" + '</th></tr>'
    total += head
    i = 0
    
    for d in data:
        unicode_line_with_spaces = get_hebrew_in_unicode(unicode_lines[i])
        volitivity_factors = find_factors(d)
        if (volitivity_factors != ""):
            style_factors = '<td style="padding:0px 15px 0px 0px;border-style:hidden;text-align:center"><div style="display:block">' + '!' + '<div style="display:none">' + volitivity_factors + '</div></div></td>'
        else:
            style_factors = '<td style="padding:0px 15px 0px 0px;border-style:hidden;text-align:center">' + ' ' + '</td>'
            
        line = '<tr>' + style_td + d[0] + '</td>'  + style_td + d[5] + '</td>' + '</td>' + style_unicode + unicode_line_with_spaces + '</td>' + style_td + d[2] + '</td>' + style_td + d[3] + '</td>' + style_td + d[7] + '</td>' + style_factors + style_td + d[13] + '</td>' + style_unicode + set_word_to_unicode(d[11]) + '</td>' + style_td + d[14] + '</td>' + style_td + d[15] + '</td>' + style_td + d[16] + '</td>' + style_td + d[17] + '</td>' + style_td + d[18] + '</td>' + style_td + '<a href="Introduction.ipynb">' + d[6] + '</a>' + '</td></tr>'
        total += line
        i += 1
    
    total += '</table>'
    
    return HTML("<strong>Psalmen 1</strong>" + total)

print_html()
```
