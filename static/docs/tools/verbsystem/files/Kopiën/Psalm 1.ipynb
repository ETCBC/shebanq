```python
import re

data = []

for line in lines:
    data_in_line = re.split(';',line)
    data.append(data_in_line)
```

```python
conversion_to_utf8 = {'>': "\u05D0",
                      'B': "\u05D1",
                      'G': "\u05D2",
                      'D': "\u05D3",
                      'H': "\u05D4",
                      'W': "\u05D5",
                      'Z': "\u05D6",
                      'X': "\u05D7",
                      'V': "\u05D8",
                      'J': "\u05D9",
                      'K': "\u05DB",
                      'L': "\u05DC",
                      'M': "\u05DE",
                      'N': "\u05E0",
                      'S': "\u05E1",
                      '<': "\u05E2",
                      'P': "\u05E4",
                      'Y': "\u05E6",
                      'Q': "\u05E7",
                      'R': "\u05E8",
                      'C': "\u05E9\u05C1",
                      'F': "\u05E9\u05C2",
                      'T': "\u05EA",
                      'p': "\u05E3",
                      'm': "\u05DD",
                      'n': "\u05DF",
                      'k': "\u05DA",
                      'y': "\u05E5",
                      '_': "-",
                      '-': "-",
                      '[': "[",
                      ']': "]",
                      '/': "/",
                      ' ': " ",
                      '\0': ""}

def set_word_to_unicode(sp):
    result = ""
    for c in sp:
        result += (conversion_to_utf8[c])
    return result
        
def analyze_phrase(subphrases):
    phrasesNew = ""
    result = ""

    if (len(subphrases) > 1):
        i = 0
        while (i < (len(subphrases) - 1)):
            phrasesNew += (set_word_to_unicode(subphrases[i]) + " ")
            i += 1
        label = subphrases[len(subphrases) - 1]
        result = label + " " + phrasesNew
    else:
        sf = re.split('-', subphrases[0])
        result = sf[1] + " -" + set_word_to_unicode(sf[0])
        
    return result

def analyze_phrases(phrases):
    final_result = ""
    i = 0
    
    for phrase in reversed(phrases):
        phrase = phrase.replace("]", "")
        final_result += "[" + analyze_phrase(phrase[1:].split()).strip() + "] "
    return final_result

def set_to_unicode(transcription, max_line_length):
    result = ""
    start_of_line = ""
    i = 0
    for c in transcription:
        if c == '[':
            break
        start_of_line += c
        i += 1
    rest_of_line = transcription[i:]
    whitespace = (max_line_length - len(transcription)) * " "
    result += whitespace + analyze_phrases(rest_of_line.split('] ')) + (''.join(reversed(start_of_line)))
    return result

def calculate_max_length(data):
    result = 0
    for d in data:
        transcription = d[19]
        if len(transcription) > result:
            result = len(transcription)
    return result
        
def analyze_transcription(data):
    unicode_lines = []
    max_line_length = calculate_max_length(data)
    for d in data:
        transcription = d[19]
        unicode_heb = set_to_unicode(transcription, max_line_length)
        unicode_lines.append(unicode_heb)
    return unicode_lines

unicode_lines = analyze_transcription(data)
```

```python
def getTranslation(d):
    result = ""
    translation_list = re.split('[ ?!]',d[20])
    translation_verb = d[21].split()
    for word in translation_list:
        if word in translation_verb:
            result += '<span class="translation_verb">' + word + '</span> '
        else:
            result += word + " "
    return result

def get_hebrew_in_unicode(line):
    unicode_line_temp = line.replace("<", "&lt;")
    unicode_line_temp2 = unicode_line_temp.replace(">", "&gt;")
    unicode_line_temp3 = unicode_line_temp2.replace(" ", "&nbsp;")
    #in order to distinguish between '< >' of tags and '< >' of constituent labels:
    unicode_line_temp4 = unicode_line_temp3.replace("!", '<span class="verb">') 
    unicode_line_temp5 = unicode_line_temp4.replace("?", "</span>")
    return (unicode_line_temp5)

def identify_verb(line, tp):
    result = ""
    phrases = line.split("[")
    i = 0
    for p in phrases:
        i += 1
        if i == 1:
            continue
        elif (p.startswith("<Pr>")) or (p.startswith("<PO>")) or (tp == "Ptcp" and p.startswith("<PC>")):
            result += '!' + "[" + p + '?' + " "
        else:
            result += "[" + p + " "
    return result

def print_html():
    total = '<table class="presentation">'
    title_head = '<tr><th></th><th>Psalm 1</th><th></th></tr>'
    head = '<tr><th>' + "Hebrew text" + '</th><th>' + '</th><th>' + "Translation" + '</th></tr>'
    total += title_head + head
    verse_label = ""
    i = 0
    
    for d in data:
        current_verse = d[0]
        if current_verse == verse_label:
            current_verse = ""
        else:
            verse_label = current_verse
        unicode_line_with_marked_verb = identify_verb(unicode_lines[i], d[2])
        unicode_line_with_spaces = get_hebrew_in_unicode(unicode_line_with_marked_verb)
        start = ""
        for c in d[19]:
            if c == '[':
                break
            start += c
        start_with_spaces = get_hebrew_in_unicode(start)

        line = '<tr><td class="unicode">' + unicode_line_with_spaces + '</td><td>' + current_verse + '</td><td>' + start_with_spaces + getTranslation(d) + '</td></tr>'
        total += line
        i += 1
    
    total += '</table>'
    
    return HTML(total)

print_html()
```

```python
def find_factors(d):
    if d[8] != "":
        if d[7] == "non-vol.":
            return "Morphological marking as " + d[8] + " overrules non-volitive default function"
        else:
            return "Morphological marking as " + d[8] + " confirms " + d[7] + " default function"
    elif d[10] != "":
        if d[10] == d[9]:
            if d[10] == "vol.pair":
                return "The clause forms a volitive pair together with its volitive daughter clause. The non-vol. default function is overridden by a volitive one."
            else:
                return "Inheritance of " + d[9] + " is possible and does indeed take place. The " + d[7] + " default function is overridden."
        else:
            if d[10] == "vol.pair":
                return "The clause could have formed a volitive pair together with its volitive daughter clause, but this is not the case. The non-vol. default function is not overridden."
            else:
                return "Inheritance of " + d[9] + " is possible, but does not take place. The " + d[7] + " default function is not overridden."
    elif d[12] != "":
        if (d[12] == "attr." or d[12] == "constit."):
            return "This 0-yiqtol clause is a " + d[12] + " clause and therefore does not fulfill its default " + d[7] + " function."
        elif d[12] == "precDrSp":
            return "This 0-yiqtol clause introduces the preceding direct speech section, which should be seen as the 0-yiqtol verb's object. Therefore the clause does not fulfill its " + d[7] + " default volitive function."
        elif d[12] == "narr":
            return "This 0-yiqtol clause is embedded in a narrative domain of communication and therefore does not fulfill its " + d[7] + "  default volitive function."
        elif d[12] == "prosp":
            return "This 0-yiqtol clause is embedded in a prospective domain of communication and therefore does not fulfill its " + d[7] + "  default volitive function."
        elif d[12] == "inhExpSu":
            return "This 0-yiqtol clause inherits its mother's explicit subject, should therefore be reanalyzed as an <X->yiqtol clause and so does not fulfill its " + d[7] + "  default volitive function."
        else:
            return "This yiqtol clause inherits the multiple-duty modifier " + set_word_to_unicode(d[11]) + " and so does not fulfill its " + d[7] + "  default volitive function."
    else:
        return ""

def assemble_default_discourse_functions(d):
    result = ""
    if (d[16] == ""):
        if (d[17] == ""):
            return ""
        else:
            return (d[14] + '-' + d[15])
    else:
        return (d[14] + '-' + d[15] + '-' + d[16])
    
def print_html():
    style_factors = ''

    total = '<table class="presentation">'
    head = '<tr><th>' + "Vs" + '</th><th>' + "\u00A7" + '</th><th>' + "Hebrew text" + '</th><th><a href="ClauseLabels.ipynb" target="_blank">' + "ClTp" + '</a></th><th><a href="DefaultFunctions.ipynb" target="_blank">' + "DefFu" + '</a></th><th><a href="Effects.ipynb" target="_blank">' + "!" + '</a></th><th><a href="FinalFunctions.ipynb" target="_blank">' + "FinalFu" + '</a></th><th><a href="MDModifier.ipynb" target="_blank">' + "MDMod" + '</a></th><th>' + "CAR" + '</th><th><a href="DiscourseFunction.ipynb" target="_blank">' + "DiscFunction" + '</a></th><th><a href="ConcordanceOfPatterns.ipynb" target="_blank">' + "#Pat" + '</a></th></tr>'
    total += head
    i = 0
    
    for d in data:
        unicode_line_with_spaces = get_hebrew_in_unicode(unicode_lines[i])
        volitivity_factors = find_factors(d)
        if (volitivity_factors != ""):
            style_factors = '<td><div class="front">!<div class="factors">' + str(volitivity_factors) + '</div></div></td>'
        else:
            style_factors = '<td>  ' + '</td>'
        df = assemble_default_discourse_functions(d)    
        line = '<tr><td>' + d[0] + '</td><td>' + d[5] + '</td><td class="unicode"><div class="front_hebrew">' + unicode_line_with_spaces + '<div class="translation">' + getTranslation(d) + '</div></div></td><td><div class="front">' + d[2] + '<div class="def_disc">' + df + '</div></div></td><td><div class="left">' + d[7] + '</div></td>' + style_factors + '<td><div class="right">' + d[13] + '</div></td><td class="unicode">' + set_word_to_unicode(d[11]) + '</td><td>' + d[3] + '</td><td>' + d[17] + '</td><td><a href="Introduction.ipynb" target="_blank">' + d[6] + '</a></td></tr>'
        total += line
        i += 1
    
    total += '</table>'
    
    return HTML("<strong>Psalmen 1</strong>" + total)

print_html()
```
