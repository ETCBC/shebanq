# The Hebrew Text

<p>In our translations and analyses of the 150 Psalms we present the consonantal Hebrew text of the Biblia Hebraica Stuttgartensia (BHS), as it has been incorporated in the linguistic database of the <abbr title="Eep Talstra Centre for Bible and Computer">ETCBC</abbr>. The Hebrew Unicode characters have been deduced from the ETCBC transcription of the Hebrew text.</p>
