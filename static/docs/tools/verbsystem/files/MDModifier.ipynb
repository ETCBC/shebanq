# Multiple-Duty Modifiers

<p>Clause-initial modifiers somehow modify the utterance made in their 'domain of governance', which is usually the clause they open. If the domain governed by a modifier consists of multiple clauses, the modifier is a multiple-duty modifier&#42;. Clauses governed by a multiple-duty modifier together constitute a single sentence and should therefore be coordinated (which in our analyses is marked by the <a href="CCR.ipynb" target="_blank">CCR</a> label <strong>Coor</strong>).</p>
<p>We distinguish between three subclasses of multiple-duty modifiers:</p>
<ol>
<li>Interrogative pronouns and phrases: <span class = "hebrew">&#1500;&#1502;&#1492;</span>, <span class = "hebrew">&#1502;&#1497;</span>, <span class = "hebrew">&#1492;-</span>, <span class = "hebrew">&#1506;&#1491; &#1506;&#1504;&#1492;</span>, etc.</li>
<li>Subordinate conjunctions (<span class = "hebrew">&#1499;&#1497;</span>, <span class = "hebrew">&#1488;&#1501;</span>, <span class="hebrew">&#1500;-&#1502;&#1506;&#1503;</span>, <span class="hebrew">&#1508;&#1503;</span>, etc.) and relative pronouns (<span class = "hebrew">&#1488;&#64298;&#1512;</span>, <span class = "hebrew">&#1494;&#64309;</span>, etc.)</li>
<li>Deictic adverbs: <span class="hebrew">&#1499;&#1492;</span>, <span class="hebrew">&#1500;&#1499;&#1503;</span>, etc.
</ol>

<p>For multiple-duty modifiers functioning as blockers of volitive functionality, see <a href="Processes.ipynb#MDM" target="_blank">here</a>.</p>
<br>
<br>

<p class="note">&#42;The term 'multiple-duty modifier' has been derived from Niccacci's term 'double-duty modifier', which might give the wrong impression that these modifying elements may govern at most two clauses.</p>
